```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

Splitting the data into train and test set.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

Performing normalization and one_hot_encoding on the numerical data before going for training

```python
X_train.columns
```

```python
X_colm= ['int.rate', 'installment', 'log.annual.inc', 'dti', 'fico',
       'days.with.cr.line', 'revol.bal', 'revol.util', 'inq.last.6mths',
       'delinq.2yrs', 'pub.rec', 'credit.policy']
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

Normalizing the numerical variable and One Hot Encoding the categorical variable

```python
tf= ColumnTransformer(transformers= [('norm',MinMaxScaler(),X_colm),
                                    ('ohe',OneHotEncoder(sparse_output=False),['purpose'])], remainder='passthrough')
```

```python
X_train_tf= tf.fit_transform (X_train)
```

```python
X_train_tf.shape
```

```python
X_test_tf= tf.transform (X_test)
```

```python
X_test_tf.shape
```

```python
X_train_tf
```

```python
X_test_tf
```

Seperating the validation dataset from the training data for pre-validation  

```python
x_val= X_train_tf[:1000]
partial_X_train= X_train_tf[1000:]

y_val= y_train[:1000]
partial_y_train= y_train[1000:]
```

```python
partial_X_train.shape
```

Applying the deep learning methods for training

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
```

Created a deep learning Architect for training with dropouts and early_stopping to overcome overfitting of the model

```python
model= Sequential([ Dense(64,activation='relu'),
                    Dropout(0.3),
                    Dense(64,activation='relu'),
                    Dropout(0.3),
                    Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop= EarlyStopping(monitor='val_loss', mode='min', patience= 10)
```

```python
history= model.fit(partial_X_train, partial_y_train, epochs=30, batch_size= 512, validation_data=(x_val,y_val), callbacks=[early_stop])
```

```python
history.history
```

Creating the plot between training and validation losses for each epochs

```python
import matplotlib.pyplot as plt
```

```python
epochs= range(1,len(history.history['loss'])+1)
```

```python
plt.plot(epochs,history.history['loss'], label= 'Training_loss')
plt.plot(epochs,history.history['val_loss'], label= 'Validation_loss')
plt.legend()
plt.show()
```

Testing of the model

```python
y_test
```

```python
results= model.evaluate(X_test_tf,y_test)
```
